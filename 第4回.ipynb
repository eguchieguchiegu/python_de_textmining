```python
import pandas as pd

df= pd.DataFrame([
    ['cola', 'S'],
    ['cola', 'M'],
    ['green tea', 'L'],
    ['milk', 'M']
], columns = ['drink', 'size'])

size2int = {'S': 0, 'M': 1, 'L': 2}

df['size'] = df['size'].map(size2int)

df.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

df['drinkLabel'] = le.fit_transform(df['drink'])

df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(binary=False)

docs = ["the cat is out of the bag", "dogs are"]

bow = vectorizer.fit_transform(docs)

bow.toarray()

print(bow)
print("---------------------")

print(vectorizer.vocabulary_)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

docs = ["the cat is out of the bag", "the dogs are"]

tfidf = vectorizer.fit_transform(docs)

import pandas as pd

vocab = vectorizer.get_feature_names_out()

pd.DataFrame(tfidf.toarray(), columns=vocab).round(2)
```

```python
from sklearn.preprocessing import MinMaxScaler

data = [[-1, 2], [-0.5, 6], [0, 10], [1, 18]]

scalar = MinMaxScaler()

data_normalized = scalar.fit_transform(data)

print(data_normalized)

import matplotlib.pyplot as plt
plt.scatter(data_normalized[:, 0], data_normalized[:, 1])
```

```python
from sklearn.preprocessing import StandardScaler

data = [[0, 10], [0, 15], [1, 20], [1, 25]]

scalar = StandardScaler()

data_standardized = scalar.fit_transform(data)

print(data_standardized)

import matplotlib.pyplot as plt
plt.scatter(data_standardized[:, 0], data_standardized[:, 1])
```
